```python
import random
import math

n = 1002
l = 20 #number of ants
alpha = 0.5
beta = 5
rEvap = 1/80 #Rate of evaporation of pheromones
pheroC = 5

edges = [[0 for i in range(n)] for j in range(n)]
phero = [[1 for i in range(n)] for j in range(n)]
ants = []

ans = []
ansFitness = 1e10
```

```python
class Ant:
    def __init__(self, myID, cur):
        self.visited = []
        self.id = myID
        self.current = int(cur)
        self.fitness = 0
    
```

```python
#Functions
def goTo(A):
    """Decides to which city ant A should go"""
    Sigma = 0
    allowed = {i for i in range(n)}
    visTemp = {x for x in A.visited}
    allowed = allowed.difference(visTemp)
    allowed = list(allowed)
    for x in allowed:
        Sigma += phero[A.current][x]**alpha * (1/edges[A.current][x]**beta)
        
    weighted = [ (phero[A.current][x]**alpha * (1/edges[A.current][x])**beta)/Sigma  for x in allowed]
    #print(allowed)
    #print(weighted)
    
    chosen = random.choices(population=allowed, weights=weighted, k = 1)
    chosen = chosen[0]
    #print(chosen)
    return int(chosen)

def Move(A):
    """Moving ant A"""
    choose = goTo(A)
    phero[A.current][choose] += pheroC/edges[A.current][choose]
    phero[choose][A.current] = phero[A.current][choose]
    A.fitness += edges[A.current][choose]
    A.current = choose
    A.visited.append(choose)
    
    

def Evaporate():
    for i in range(n):
        for j in range(i+1, n):
            phero[i][j] *= (1-rEvap)
            phero[i][j] = max(phero[i][j], 0.2)
            phero[j][i] = phero[i][j]
            
            
            
def fitness(a):
    #returns the fitness of the permutation a
    print("Guaranteed ")
    fit = int(0)
    for i in range(0, len(a)-1):
        fit = fit+edges[a[i]][a[i+1]]
    fit = fit+edges[a[len(a)-1]][a[0]] #back to the first node
    
    return fit
```

```python
#Preprocess
for i in range (n):
    for j in range(n):
        if i != j:
            phero[i][j] = 1
```

```python
ages = 1000
history = []
for _ in range(ages):
    ants = []
    if _%1 == 0 and _ !=0:
        print(_, ansFitness)
        history.append(ansFitness)
    for i in range(l):
        firstPos = random.choice(range(n))
        newAnt = Ant(myID = i, cur = firstPos)
        newAnt.visited.append(firstPos)
        ants.append(newAnt)
        
    for i in range(n-1): #Every ant must move n-1 time
        
        for j in range(l):
            Move(ants[j]);
    
    for j in range(l):
        ants[j].fitness+= edges[ants[j].current][ants[j].visited[0]] #back to the first one
    
    for i in range(l):
        if ants[i].fitness < ansFitness:
            ansFitness = ants[i].fitness
            ans = ants[i].visited
    
    
    
    Evaporate()
```
